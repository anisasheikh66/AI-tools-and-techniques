# Predict Proﬁt for food truck
Problem statement : 
In this project, we are implementing linear regression with one variable to predict proﬁts for a food truck.The ﬁle ex1data1.txt contains the dataset for our linear regression problem. The ﬁrst column is the population of a city and the second column is the proﬁt of a food truck in that city. A negative value for proﬁt indicates a loss. 
Now we have to predict proﬁt for given population city(including which is not traverse by our food truck). 

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd  #to work on the database
```

```python
#reading data
data=pd.read_csv("profitdataset.txt")
print(data.shape)
```

```python
x=data['population'].values
y=data['profit'].values
```

```python
#plot data values
plt.scatter(x,y,40,c='r',label='scatter_data',marker='x')
plt.title("food truck profit Prediction")
plt.xlabel("population")
plt.ylabel("profit")
plt.grid(True,color='k')
plt.show()
```

```python
#mean values
mean_x=np.mean(x)
mean_y=np.mean(y)
n=len(x)
mean_x
```

```python
#y=mx+c, to get the values m&c
num=0
den=0
for i in range(n):
    num=num+((x[i]-mean_x)*(y[i]-mean_y))
    den=den+((x[i]-mean_x)**2)
m=num/den
#c=y-m*x
c=mean_y-(m*mean_x)
print(m,c)
```

```python
#calculate y_predicate
y_pred=np.array([]) #to create an empty array

for i in range(n):
    p=(m*x[i])+c  #y=mx+c
    y_pred=np.append(y_pred,np.array([p]),axis=0)
print(y_pred)
```

```python
#plot the regression line
plt.plot(x,y_pred,color='m',label="regression_line")
#plot the actual data
plt.scatter(x,y,40,c='g',label="actual_data",marker="o")
plt.scatter(x,y_pred,40,c='k',label="predicted_data",marker="x")
plt.xlabel("population")
plt.ylabel("profit")
plt.legend()
plt.show()
```

```python
#how good our fit line or regression line is by using r square 

nu=0
de=0
for i in range(n):
    nu=nu+((y_pred[i]-mean_y)**2) #predicted values of y
    de=de+((y[i]-mean_y)**2)     #actual values of y
r2=nu/de
r2
```
