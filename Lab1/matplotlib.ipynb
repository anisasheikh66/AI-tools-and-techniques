# Matplot Library

# linear graph

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x1=[2,4,6]
y1=[3,3,3]
x2=[2,4,6]
y2=[4,8,12]
plt.plot(x1,y1,label="slope=0")
plt.plot(x2,y2,label="slope>0")
plt.xlabel("time")
plt.ylabel("displacement")
plt.title("Graphs of Motion")
plt.legend()
plt.grid()
```

# Bar graph

```python
plt.bar([6,7,8,9,10],[90,70,80,50,40],width=0.5,align='center',label="girls",color="y")
plt.bar([6,7,8,9,10],[100,90,80,70,60],width=0.3,align='edge',label="boys")
plt.legend()
plt.xlabel("class")
plt.ylabel("No.of.students")
plt.title("Boys and Girls in each Class")
plt.show()
```

# Histogram

```python
age=[27,26,7,79,54,58,22,36,44,33,39,16,65,2,0]
ranges=[0,15,30,45,60,75,90]
plt.hist(age,ranges,label="population",histtype="bar",rwidth=0.5)

plt.title("Demographic trends in India")
plt.xlabel("Age range")
plt.ylabel("Percentage")

plt.legend()
plt.show()
```

# Scatter plot

```python
weight=[52.3,49.5,63.8,52.1,71.4,63.6]
height=[5.1,5.2,5.3,5.4,5.5,5.6]
plt.scatter(weight,height,label='BMI',s=25,marker='*',color='k')

plt.title("Weight vs Height plot")
plt.xlabel("Weight in Kgs")
plt.ylabel("Height in mtrs")
plt.legend()
plt.show()
```

# stack plot

```python
year=[1900,1920,1940,1960,1980,2000]
agricultre=[500,1000,1300,1500,2000,3000]
industry=[50,70,200,250,300,500]
municipal=[50,50,50,100,250,400]
reservoir=[30,30,30,40,45,50]
labels=['agricultre','industry','municipal use','reservoir losses']
plt.stackplot(year,agricultre,industry,municipal,reservoir,labels=labels,colors=['b','r','y','g'])

plt.title("Water Usage per sector")
plt.xlabel("Year")
plt.ylabel("cubic kilometers per year")
plt.legend()
plt.show
```

# Pie Chart

```python
slices=[22,10,7,20,33,8]
expense=['Food','Medicine',"Entertainment",'saving',"housing","Others"]
colors=['c','g','m','y','b','r']
plt.pie(slices,labels=expense,colors=colors,startangle=0)
plt.title("Monthly Expenditure of a family")
plt.show()
```
